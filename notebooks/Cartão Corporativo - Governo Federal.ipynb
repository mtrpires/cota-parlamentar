```python
import pandas
import numpy
```

```python
base = pandas.read_csv('governo_federal_2015-2016.csv', sep='\t', encoding='latin_1', error_bad_lines=False, warn_bad_lines=True, low_memory=False)
```

```python
base
```

```python
base.describe()
```

```python
base.pivot_table(index=u'Nome Órgão Superior', values=u'Valor Transação', aggfunc=numpy.sum)
```

```python
base[u'Valor Transação'] = base[u'Valor Transação'].str.replace('.', '')
```

```python
base[u'Valor Transação'] = base[u'Valor Transação'].str.replace(',', '.')
```

```python
base[u'Valor Transação'] = base[u'Valor Transação'].convert_objects(convert_numeric=True)
```

```python
base.pivot_table(index=u'Nome Órgão Superior', values=u'Valor Transação', aggfunc=numpy.sum)
```
